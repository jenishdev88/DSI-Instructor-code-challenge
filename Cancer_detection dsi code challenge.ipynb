```python
#split the dataset in to independent and dependent data set

x.iloc[:,1]
k = x.iloc[:,2:31].values #detects the features that causes cancer
l = x.iloc[:,1].values    #says whether the patient has cancer
```

```python
from sklearn.preprocessing import LabelEncoder
encoder_x = LabelEncoder()
x.iloc[:,1] = encoder_x.fit_transform(x.iloc[:,1].values)
```

```python
#split the data set to training and test
from sklearn.model_selection import train_test_split
k_train,k_test,l_train,l_test =train_test_split(k,l,test_size = 0.25, random_state = 0)
```

```python
#scaling the data to bring it to same level of magnitude(feature scaling)
from sklearn.preprocessing import StandardScaler
scaling = StandardScaler()
k_train = scaling.fit_transform(k_train)   #transform the feature 
k_test = scaling.fit_transform(k_test)
```

```python
# Create a function for models

def model(k_train, l_train):

    from sklearn.linear_model import LogisticRegression
    regr = LogisticRegression(random_state=0)
    regr.fit(k_train,l_train)

    from sklearn.tree import DecisionTreeClassifier
    tree = DecisionTreeClassifier(criterion = 'entropy', random_state=0)
    tree.fit(k_train,l_train)

# model accuracy

    print('[0]Accuracy of logistic regression is:', regr.score(k_train,l_train))
    print('[1]Accuracy of Decision tree classifier is:', tree.score(k_train,l_train))
    
    return regr,tree
        
        
```

```python
models= model(k_train, l_train)
```

```python
#Test the model on testing data using confusion matrix
 
from sklearn.metrics import confusion_matrix
conf = confusion_matrix(l_test,models[0].predict(k_test))
print(conf)

```

```python
#Find the accuracy of the testing model
for i in range(len(models)):
    conf = confusion_matrix(l_test,models[i].predict(k_test))

    tp = conf[0][0]  # true positive value from the conf
    tn = conf[1][1]  # true negative value from the conf
    fn = conf[0][1]  # false negative value from the conf
    fp = conf[1][0]  # false positive value from the conf
    print(conf)
    print("testing accuracy=", (tp +tn)/(tp+tn+fp+fn))

```
